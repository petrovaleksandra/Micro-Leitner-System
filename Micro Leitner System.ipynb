# Simulation Micro Leinter System

The simulation is done by running a number of sessions that would be on the app

```python
n_sessions = input()
n_sessions = int(n_sessions)
```

For the puprpose of the simulation a dictionary from English to French is used as an example

```python
cards = {'cat':'chat', 'dog':'chien', 'fish':'poison', 'rabbit':'lapin', 'mouse':'souris', 'parrot':'perroquet', 
         'hamster':'hamster', 'horse':'cheval', 'cow':'vache', 'duck':'canard'}
```

The code consists of the global decks where the changes that happen within sessions remain until the next one. In the begining the first deck is initialized with all of the information from the dictionary

```python
import queue
```

```python
g_deck1 = queue.Queue()
has_moved_globaly = {}

for key in cards:
    g_deck1.put([key, cards.get(key)])
    has_moved_globaly[key] = False
```

```python
g_decks = [g_deck1]
g_deck2 = queue.Queue()
g_decks.append(g_deck2)
g_deck3 = queue.Queue()
g_decks.append(g_deck3)
g_deck4 = queue.Queue()
g_decks.append(g_deck4)
g_deck5 = queue.Queue()
g_decks.append(g_deck5)
```

To be able to track if a card has advanced or been brought back during a session a hash table is used for each card to track if the card has already moved from the session deck or not. If it hasn't then it should also be moved globaly, if it has then the global deck has been adjusted already.

```python
x = input()
x = int(x)
p = input()
p = int(p)
```

```python
accepted_answers = {'y', 'n', 'quit'}
```

```python
for s in range(n_sessions):
    print ('Session ' + str(s+1))
    print ('\n')
    quit = False
    
    #Each session starts by testing out the P amount cards from the final deck
    for _ in range(p):
        if(not g_decks[-1].empty()):
            [eng, fre] = g_decks[-1].get()
            print(eng + 'is picked from final deck')
            print('The answer is: ' + fre)
            print('Did you know the correct answer for 1? (y/n/quit)')
            answer = input()
            while(answer not in accepted_answers):
                print('Please only type in y, n or quit')
                answer = input()
            if(answer == 'n'):
                g_decks[-2].put([eng, fre])
            elif(answer == ' quit'):
                break
    s_decks = g_decks.copy()
    has_moved_globaly = dict.fromkeys(has_moved_globaly, False)
    
    #The remainder of the decks in each session are iterated from here
    while(True):      
        #If all decks besides the final one are empty here we loop through all the cards in the final one
        if(s_decks[0].empty() and s_decks[1].empty() and s_decks[2].empty() and s_decks[3].empty()):
            [eng, fre] = g_decks[4].get()
            print(eng + ' is picked from final deck')
            print('The answer is: ' + fre)
            print('Did you know the correct answer for 1? (y/n/quit)')
            answer = input()
            while(answer not in accepted_answers):
                print('Please only type in y, n or quit')
                answer = input()
            if(answer == 'n'):
                s_decks[4].put([eng, fre])
            elif(answer == 'quit'):
                quit = True
                break
        
        #Going through X, X/2, X/4, X/8 from each of the decks. A card gets picked from another deck only when the loop has passed more than 1/2, 3/4 or 7/8 iterations
        for i in range(x):
            if(not s_decks[0].empty()):
                [eng, fre] = s_decks[0].get()
                print(eng + ' is picked from deck 1')
                print('The answer is: ' + fre)
                print('Did you know the correct answer for 1? (y/n/quit)')
                answer = input()
                while(answer not in accepted_answers):
                    print('Please only type in y, n or quit')
                    answer = input()
                if(answer == 'y'):
                    s_decks[1].put([eng, fre])
                    if(has_moved_globaly[eng] == False):
                        s_decks[0].get()
                        g_decks[1].put([eng,fre])
                elif(answer == 'n'):
                    s_decks[0].put([eng, fre])
                    if(has_moved_globaly[eng] == False):
                        s_decks[0].get()
                        g_decks[0].put([eng,fre])
                elif(answer == 'quit'):
                    quit = True
                    break
            if(i > x/2):
                if(not s_decks[1].empty()):
                    [eng, fre] = s_decks[1].get()
                    print(eng + ' is picked from deck 2')
                    print('The answer is: ' + fre)
                    print('Did you know the correct answer for 1? (y/n/quit)')
                    answer = input()
                    while(answer not in accepted_answers):
                        print('Please only type in y, n or quit')
                        answer = input()
                    if(answer == 'y'):
                        s_decks[1].put([eng, fre])
                        if(has_moved_globaly[eng] == False):
                            s_decks[1].get()
                            g_decks[2].put([eng,fre])
                    elif(answer == 'n'):
                        s_decks[0].put([eng, fre])
                        if(has_moved_globaly[eng] == False):
                            s_decks[1].get()
                            g_decks[0].put([eng,fre])
                    elif(answer == 'quit'):
                        quit = True
                        break
            if(i > x-x/4):
                if(not s_decks[2].empty()):
                    [eng, fre] = s_decks[2].get()
                    print(eng + ' is picked from deck 3')
                    print('The answer is: ' + fre)
                    print('Did you know the correct answer for 1? (y/n/quit)')
                    answer = input()
                    while(answer not in accepted_answers):
                        print('Please only type in y, n or quit')
                        answer = input()
                    if(answer == 'y'):
                        s_decks[3].put([eng, fre])
                        if(has_moved_globaly[eng] == False):
                            s_decks[2].get()
                            g_decks[3].put([eng,fre])
                    elif(answer == 'n'):
                        s_decks[1].put([eng, fre])
                        if(has_moved_globaly[eng] == False):
                            s_decks[2].get()
                            g_decks[1].put([eng,fre])
                    elif(answer == 'quit'):
                        quit = True
                        break
            if(i > x-x/8):
                if(not s_decks[3].empty()):
                    [eng, fre] = s_decks[3].get()
                    print(eng + ' is picked from deck 4')
                    print('The answer is: ' + fre)
                    print('Did you know the correct answer for 1? (y/n/quit)')
                    answer = input()
                    while(answer not in accepted_answers):
                        print('Please only type in y, n or quit')
                        answer = input()
                    if(answer == 'y'):
                        s_decks[4].put([eng, fre])
                        if(has_moved_globaly[eng] == False):
                            s_decks[3].get()
                            g_decks[4].put([eng,fre])
                    elif(answer == 'n'):
                        s_decks[2].put([eng, fre])
                        if(has_moved_globaly[eng] == False):
                            s_decks[3].get()
                            g_decks[2].put([eng,fre])
                    elif(answer == 'quit'):
                        quit = True
                        break
        if(quit == True):
            print('\n')
            break
```
