# Simulation Micro Leinter System

For the puprpose of the simulation a dictionary from English to French is used as an example

```python
cards = {'cat':'chat', 'dog':'chien', 'fish':'poison', 'rabbit':'lapin', 'mouse':'souris', 'parrot':'perroquet', 
         'hamster':'hamster', 'horse':'cheval', 'cow':'vache', 'duck':'canard'}
n_cards = len(cards)
print(n_cards)
```

The code consists of the global decks where the changes that happen within sessions remain until the next one. In the begining the first deck is initialized with all of the information from the dictionary

```python
g_decks = {}
has_moved_globaly = {}
```

To be able to track if a card has advanced or been brought back during a session a hash table is used for each card to track if the card has already moved from the session deck or not. If it hasn't then it should also be moved globaly, if it has then the global deck has been adjusted already.

```python
x = input()
x = int(x)
p = input()
p = int(p)
y = input()
y = int(y)
```

```python
accepted_answers = {'y', 'n', 'quit', 'session'}
in_deck1 = 0
```

```python
for s in range(1000):
    print ('Session ' + str(s+1))
    print ('\n')
    session = False
    quit = False
    
    #Add up to y new words 
    if(not bool(cards) ):
        for j in range(y-in_deck1):
            g_decks[('card'+str(n_cards)+' '+'answ'+str(n_cards))] = 1
            has_moved_globaly[('card'+str(n_cards)+' '+'answ'+str(n_cards))] = False
            n_cards+=1
        in_deck1=y
    else:
        for key in cards:
            g_decks[(key+' '+cards.get(key))] = 1
            has_moved_globaly[(key+' '+cards.get(key))] = False
            in_deck1+=1
            if(in_deck1==y):
                break
        if(in_deck1 < y):
            for j in range(y-in_deck1):
                g_decks[('card'+str(n_cards)+' '+'answ'+str(n_cards))] = 1
                has_moved_globaly[('card'+str(n_cards)+' '+'answ'+str(n_cards))] = False
                n_cards+=1

    s_deck1 = []
    s_deck2 = []
    s_deck3 = []
    s_deck4 = []
    s_deck5 = []
    s_decks = [s_deck1, s_deck2, s_deck3, s_deck4, s_deck5]
    
    for key in g_decks:
        s_decks[g_decks[key]-1].append(key)
    
    #Each session starts by testing out the P amount cards from the final deck
    for _ in range(p):
        if(s_decks[-1]):
            [eng, fre] = s_decks[-1].pop(0).split(' ')
            print(eng + 'is picked from final deck')
            print('The answer is: ' + fre)
            print('Did you know the correct answer for it? (y/n/quit/session)')
            answer = input()
            while(answer not in accepted_answers):
                print('Please only type in y, n, quit or session')
                answer = input()
            if(answer == 'y'):
                del g_decks[(eng+' '+fre)]
            elif(answer == 'n'):
                s_decks[-2].append((eng+' '+fre))
            elif(answer == ' quit'):
                quit = True
                break
            else:
                break

    has_moved_globaly = dict.fromkeys(has_moved_globaly, False)
    
    #The remainder of the decks in each session are iterated from here
    while(True):      
        #If all decks besides the final one are empty here we loop through all the cards in the final one
        if(not s_decks[0] and not s_decks[1] and not s_decks[2] and not s_decks[3]):
            [eng, fre] = s_decks[-1].pop(0).split(' ')
            print(eng + 'is picked from final deck')
            print('The answer is: ' + fre)
            print('Did you know the correct answer for it? (y/n/quit/session)')
            answer = input()
            while(answer not in accepted_answers):
                print('Please only type in y, n, quit or session')
                answer = input()
            if(answer == 'n'):
                s_decks[3].append((eng+' '+fre))
            elif(answer == 'session'):
                s_decks[-1].append((eng+' '+fre))
                break
            elif(answer == 'quit'):
                s_decks[-1].append((eng+' '+fre))
                quit = True
                break
        
        #Going through X, X/2, X/4, X/8 from each of the decks. A card gets picked from another deck only when the loop has passed more than 1/2, 3/4 or 7/8 iterations
        for i in range(x):
            if(s_decks[0]):
                [eng, fre] = s_decks[0].pop(0).split(' ')
                print(eng + ' is picked from deck 1')
                print('The answer is: ' + fre)
                print('Did you know the correct answer for it? (y/n/quit/session)')
                answer = input()
                while(answer not in accepted_answers):
                    print('Please only type in y, n, quit or session')
                    answer = input()
                if(answer == 'y'):
                    s_decks[1].append((eng+' '+fre))
                    if(has_moved_globaly[(eng+' '+fre)] == False):
                        g_decks[(eng+' '+fre)] = 2
                        has_moved_globaly[(eng+' '+fre)] = True
                elif(answer == 'n'):
                    s_decks[0].append((eng+' '+fre))
                elif(answer == 'session'):
                    s_decks[0].append((eng+' '+fre))
                    session = True
                    break
                elif(answer == 'quit'):
                    s_decks[0].append((eng+' '+fre))
                    quit = True
                    break
            if(i >= x/2):
                if(s_decks[1]):
                    [eng, fre] = s_decks[1].pop(0).split(' ')
                    print(eng + ' is picked from deck 2')
                    print('The answer is: ' + fre)
                    print('Did you know the correct answer for it? (y/n/quit/session)')
                    answer = input()
                    while(answer not in accepted_answers):
                        print('Please only type in y, n, quit or session')
                        answer = input()
                    if(answer == 'y'):
                        s_decks[2].append((eng+' '+fre))
                        if(has_moved_globaly[(eng+' '+fre)] == False):
                            g_decks[(eng+' '+fre)] = 3
                            has_moved_globaly[(eng+' '+fre)] = True
                    elif(answer == 'n'):
                        s_decks[0].append((eng+' '+fre))
                        if(has_moved_globaly[(eng+' '+fre)] == False):
                            g_decks[(eng+' '+fre)] = 1
                            has_moved_globaly[(eng+' '+fre)] = True
                    elif(answer == 'session'):
                        s_decks[1].append((eng+' '+fre))
                        session = True
                        break
                    elif(answer == 'quit'):
                        s_decks[1].append((eng+' '+fre))
                        quit = True
                        break
            if(i >= x-x/4):
                if(s_decks[2]):
                    [eng, fre] = s_decks[2].pop(0).split(' ')
                    print(eng + ' is picked from deck 3')
                    print('The answer is: ' + fre)
                    print('Did you know the correct answer for it? (y/n/quit/session)')
                    answer = input()
                    while(answer not in accepted_answers):
                        print('Please only type in y, n, quit or session')
                        answer = input()
                    if(answer == 'y'):
                        s_decks[3].append((eng+' '+fre))
                        if(has_moved_globaly[(eng+' '+fre)] == False):
                            g_decks[(eng+' '+fre)] = 4
                            has_moved_globaly[(eng+' '+fre)] = True
                    elif(answer == 'n'):
                        s_decks[1].append((eng+' '+fre))
                        if(has_moved_globaly[(eng+' '+fre)] == False):
                            g_decks[(eng+' '+fre)] = 2
                            has_moved_globaly[(eng+' '+fre)] = True
                    elif(answer == 'session'):
                        s_decks[2].append((eng+' '+fre))
                        session = True
                        break
                    elif(answer == 'quit'):
                        s_decks[2].append((eng+' '+fre))
                        quit = True
                        break
            if(i >= x-x/8):
                if(s_decks[3]):
                    [eng, fre] = s_decks[3].pop(0).split(' ')
                    print(eng + ' is picked from deck 4')
                    print('The answer is: ' + fre)
                    print('Did you know the correct answer for it? (y/n/quit/session)')
                    answer = input()
                    while(answer not in accepted_answers):
                        print('Please only type in y, n, quit or session')
                        answer = input()
                    if(answer == 'y'):
                        s_decks[4].append((eng+' '+fre))
                        if(has_moved_globaly[(eng+' '+fre)] == False):
                            g_decks[(eng+' '+fre)] = 5
                            has_moved_globaly[(eng+' '+fre)] = True
                    elif(answer == 'n'):
                        s_decks[2].append((eng+' '+fre))
                        if(has_moved_globaly[(eng+' '+fre)] == False):
                            g_decks[(eng+' '+fre)] = 3
                            has_moved_globaly[(eng+' '+fre)] = True
                    elif(answer == 'session'):
                        s_decks[3].append((eng+' '+fre))
                        session = True
                        break
                    elif(answer == 'quit'):
                        s_decks[3].append((eng+' '+fre))
                        quit = True
                        break
                        
        if(session == True):
            print('\n')
            break
        
        if(quit == True):
            break
            
    if(quit == True):
        break
```
